```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

<h2>Importing data.json file from the given url</h2>

```python
from urllib.request import urlopen
url = "https://raw.githubusercontent.com/localytics/data-viz-challenge/master/data.json"
jsonurl = urlopen(url)
text = json.loads(jsonurl.read())
```

```python
text
```

<b>Normalizing the dictionary data in text variable</b>

```python
from pandas.io.json import json_normalize
df=json_normalize(text['data'])
```

<b>Here we are selecting columns that are useful for us and creating new dataframe to manipulate the data</b>

```python
new_frame=df.drop(['category','event_name','location.latitude','location.longitude','marital_status','client_time','location.zip_code'],axis=1)
```

```python
new_frame.head()
```

<b>As discussed before since the amount column has many NaN values we are simply replacing it with 0 so that it is easy to manipulate the columns data</b>

```python
new_frame['amount']=new_frame['amount'].fillna(0)
```

<b> Since we need data only for the gender "Female" denoted here by F so filtering out rest of the genders by the below code</b>

```python
new_frame=new_frame[new_frame['gender']=='F']
```

```python
new_frame.head()
```

```python
new_frame['location.state'].unique()
```

<b>Since, we also need data filtered by the state California, so further filtration is performed by the below code</b>

```python
new_frame=new_frame[new_frame['location.state']=='CA']
```

```python
new_frame.info()
```

```python
#new_frame['amount']=new_frame['amount'].fillna(0)
```

```python
new_frame.tail()
```

<b>In the below code we are aggregating data at the amount level using sum and storing it in f1 as series object</b>

```python
f1=new_frame.groupby(['gender','age','device','Date']).sum()
```

<b>In the below code we are counting the number of entries per session id</b>

```python
f2=new_frame.groupby(['gender','age','device','Date'])['session_id'].count()
```

```python
print(f2)
```

<b>Creating dataframe by combining the below series object</b>

```python
f3=pd.concat([f1, f2], axis=1).reset_index()
```

```python
f3
```

```python
total_events=f3.rename(index=str, columns={"amount": "sum", "session_id": "count"})
```

```python
total_events
```

```python
total_events=total_events.drop('gender',axis=1)
```

```python
new_order = [0,1,2,4,3]
total_events = total_events[total_events.columns[new_order]]
print(total_events)  
```

<b>Total events id the dataframe that has been aggregated at amount level and counted by no of enteries, filtered by the gender 'F' and state "california". Grouped by age , device type, Date</b>

```python
total_events.to_csv('total_events.csv',index=False)
```
